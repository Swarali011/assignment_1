```python
import pandas as pd 
```

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
x=np.array([10,20,30,40,50,60,70,80,90,11])
y=np.array([11,22,33,44,55,66,77,88,99,19])
z=np.array([15,26,36,47,57,68,79,84,96,12])
```

# line plot 


```python
plt.title('line plot')
plt.xlabel('First value')
plt.ylabel('second value')
plt.xticks(range(1,90,5))
plt.yticks(range(1,90,5))
plt.plot(x,y,marker='o',mfc='red',color='black',ls='--',ms=5);
plt.grid()
```

```python

plt.title('line plot')
plt.xlabel('First value')
plt.ylabel('second value')
plt.xticks(range(1,90,5))
plt.yticks(range(1,90,5))
plt.plot(x,z,marker='H',mfc='red',color='black',ls='--',ms=5);

plt.grid()
#marker=> >,<,*,H,o,^
```

```python

plt.title('line plot')
plt.xlabel('First value')
plt.ylabel('second value')
plt.xticks(range(1,90,5))
plt.yticks(range(1,90,5))
plt.plot(x,z,marker='o',mfc='red',color='black',ls='--',ms=5,label='2006');
plt.plot(x,y,marker='o',mfc='yellow',color='black',ls='--',ms=5,label='2024');
plt.grid()
plt.legend();
#marker=> >,<,*,H,o,^
```

```python
#sub plot
plt.figure(figsize=(16,9))
plt.subplot(2,2,1)
plt.title('line plot')
plt.xlabel('First value')
plt.ylabel('second value')
plt.xticks(range(1,90,5))
plt.yticks(range(1,90,5))
plt.plot(x,y,marker='o',mfc='red',color='black',ls='--',ms=5);
plt.grid()

plt.subplot(2,2,2)
plt.title('line plot')
plt.xlabel('First value')
plt.ylabel('second value')
plt.xticks(range(1,90,5))
plt.yticks(range(1,90,5))
plt.plot(x,z,marker='H',mfc='red',color='black',ls='--',ms=5);

plt.grid()
```

# bar plot 

```python
courses=['java','python','kotlin','dart','R']
student=[34,56,78,89,76]
plt.bar(courses,student,color='red',edgecolor='black',width=0.3,linewidth=3,alpha=0.5);
plt.title('bar plot')
plt.xlabel('courses')
plt.ylabel('student')
plt.grid()
```
